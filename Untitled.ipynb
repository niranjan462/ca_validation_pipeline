```python
import pandas as pd
import numpy as np
import sqlite3
from sqlalchemy import create_engine
import os

print("✅ All imports working!")
print("Pandas version:", pd.__version__)
```

```python
os.makedirs('data', exist_ok=True)
engine = create_engine('sqlite:///data/ca_data.db')
print("✅ DB ready at data/ca_data.db")
```

```python
df_nse = pd.read_csv('nse_ca.csv')
print("📊 NSE Data Loaded!")
print("Shape (rows, cols):", df_nse.shape)
print("\n📋 Column names:")
print(df_nse.columns.tolist())
print("\n🔍 First 3 rows:")
print(df_nse.head(3))
df_nse.to_sql('ca_source_nse', engine, if_exists='replace', index=False)
print("\n✅ Saved to database!")

```
